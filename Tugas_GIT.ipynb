```python
data_panen = {
    'lokasi1': {
        'nama_lokasi': 'Kebun A',
        'hasil_panen': {
            'padi': 1200,
            'jagung': 800,
            'kedelai': 500
        }
    },
    'lokasi2': {
        'nama_lokasi': 'Kebun B',
        'hasil_panen': {
            'padi': 1500,
            'jagung': 900,
            'kedelai': 450
        }
    },
    'lokasi3': {
        'nama_lokasi': 'Kebun C',
        'hasil_panen': {
            'padi': 1100,
            'jagung': 750,
            'kedelai': 600
        }
    },
    'lokasi4': {
        'nama_lokasi': 'Kebun D',
        'hasil_panen': {
            'padi': 1300,
            'jagung': 850,
            'kedelai': 550
        }
    },
    'lokasi5': {
        'nama_lokasi': 'Kebun E',
        'hasil_panen': {
            'padi': 1400,
            'jagung': 950,
            'kedelai': 480
        }
    }
}

print(data_panen)

for lokasi, i in data_panen.items():
    print(f"{i['nama_lokasi']}:")
    for jenis, hasil in i['hasil_panen'].items():
        print(f"  {jenis}: {hasil} kg")
print("-------------------------------------")
hsljagunglks2 = data_panen['lokasi2']['hasil_panen']['jagung']
print(f"Hasil panen jagung dari lokasi2: {hsljagunglks2} kg")
print("-------------------------------------")
nmalks3 = data_panen['lokasi3']['nama_lokasi']
print(f"Nama lokasi dari lokasi3: {nmalks3}")

hslpdi = {lokasi: i['hasil_panen']['padi'] for lokasi, i in data_panen.items()}
hslkdelai = {lokasi: i['hasil_panen']['kedelai'] for lokasi, i in data_panen.items()}
print("Hasil panen padi dari masing masing lokasi:")
for lokasi, padi in hslpdi.items():
    print(f"{lokasi}: {padi} kg")  
print("\nHasil panen kedelai dari masing masing lokasi:")
for lokasi, kedelai in hslkdelai.items():
    print(f"{lokasi}: {kedelai} kg")
print("-------------------------------------")
for lokasi, i in data_panen.items():
    padi = i['hasil_panen']['padi']
    jagung = i['hasil_panen']['jagung']
    if padi > 1300 or jagung > 800:
        print(f"{i['nama_lokasi']} memerlukan perhatian khusus.")
    else:
        print(f"{i['nama_lokasi']} dalam kondisi baik.")
```
